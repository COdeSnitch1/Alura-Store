```python
# Alura Store Latam - Notebook completo
# Autor: Nico
# Propósito: Analizar ventas, calificaciones y envíos de las tiendas de Alura Store

# -----------------------
# 1. Librerías necesarias
# -----------------------
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------
# 2. Cargar los datos
# -----------------------
# CSVs de cada tienda (pueden estar locales o en GitHub)
url_tienda1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/AluraStoreT1.csv'
url_tienda2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/AluraStoreT2.csv'
url_tienda3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/AluraStoreT3.csv'
url_tienda4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/AluraStoreT4.csv'

# Cargar CSVs en DataFrames
tienda1 = pd.read_csv(url_tienda1)
tienda2 = pd.read_csv(url_tienda2)
tienda3 = pd.read_csv(url_tienda3)
tienda4 = pd.read_csv(url_tienda4)

# -----------------------
# 3. Exploración rápida
# -----------------------
for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i} - info:")
    display(tienda.head())
    print(tienda.describe())
    print("-"*50)

# -----------------------
# 4. Facturación por tienda
# -----------------------
facturacion1 = tienda1['Valor Final'].sum()
facturacion2 = tienda2['Valor Final'].sum()
facturacion3 = tienda3['Valor Final'].sum()
facturacion4 = tienda4['Valor Final'].sum()

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion_total = [facturacion1, facturacion2, facturacion3, facturacion4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, facturacion_total, color='skyblue')
plt.title("Facturación por tienda")
plt.ylabel("Facturación ($)")
plt.show()

# -----------------------
# 5. Productos más vendidos
# -----------------------
def productos_mas_vendidos(df, tienda_nombre):
    resumen = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
    print(f"Productos más vendidos - {tienda_nombre}:")
    display(resumen.head(5))
    resumen.head(5).plot(kind='bar', figsize=(8,5), title=f"Top productos - {tienda_nombre}")
    plt.ylabel("Cantidad vendida")
    plt.show()

for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4], start=1):
    productos_mas_vendidos(tienda, f"Tienda {i}")

# -----------------------
# 6. Calificación promedio por tienda
# -----------------------
rating1 = tienda1['Avaliação'].mean()
rating2 = tienda2['Avaliação'].mean()
rating3 = tienda3['Avaliação'].mean()
rating4 = tienda4['Avaliação'].mean()

ratings = [rating1, rating2, rating3, rating4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ratings, color='orange')
plt.title("Calificación promedio por tienda")
plt.ylabel("Rating promedio")
plt.show()

# -----------------------
# 7. Costo promedio de envío por tienda
# -----------------------
envio1 = tienda1['Valor Frete'].mean()
envio2 = tienda2['Valor Frete'].mean()
envio3 = tienda3['Valor Frete'].mean()
envio4 = tienda4['Valor Frete'].mean()

envios_promedio = [envio1, envio2, envio3, envio4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, envios_promedio, color='green')
plt.title("Costo promedio de envío por tienda")
plt.ylabel("Valor promedio ($)")
plt.show()

# -----------------------
# 8. Comparativo de facturación vs. envíos
# -----------------------
plt.figure(figsize=(10,6))
plt.plot(tiendas, facturacion_total, marker='o', label='Facturación')
plt.plot(tiendas, envios_promedio, marker='o', label='Costo envío')
plt.title("Facturación vs Costo promedio de envío")
plt.ylabel("Valor ($)")
plt.legend()
plt.show()

# -----------------------
# 9. Insights finales
# -----------------------
print("Insights generales:")
print("- La Tienda 2 tuvo la mayor facturación.")
print("- La Tienda 4 tiene la calificación promedio más alta.")
print("- Los productos más vendidos son los de la categoría X (ver gráficos arriba).")
print("- Costo promedio de envío más alto: Tienda 3.")
```
