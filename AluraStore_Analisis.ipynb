```python
# ==============================
# CHALLENGE ALURA STORE LATAM
# ==============================

# 1 Importar librerías
import pandas as pd
import matplotlib.pyplot as plt

# 2 Cargar los datos de las tiendas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# 3 Revisar datos
print("Datos tienda 1")
print(tienda1.head())

# ==============================
# FACTURACIÓN POR TIENDA
# ==============================

facturacion1 = tienda1["Precio"].sum()
facturacion2 = tienda2["Precio"].sum()
facturacion3 = tienda3["Precio"].sum()
facturacion4 = tienda4["Precio"].sum()

print("\nFacturación por tienda")
print("Tienda 1:", facturacion1)
print("Tienda 2:", facturacion2)
print("Tienda 3:", facturacion3)
print("Tienda 4:", facturacion4)

# ==============================
# VENTAS POR CATEGORÍA
# ==============================

categoria1 = tienda1.groupby("Categoria del Producto")["Precio"].sum()

print("\nVentas por categoría - Tienda 1")
print(categoria1)

# ==============================
# PRODUCTOS MÁS VENDIDOS
# ==============================

productos_mas_vendidos = tienda1["Producto"].value_counts().head(10)

print("\nProductos más vendidos")
print(productos_mas_vendidos)

# ==============================
# CALIFICACIÓN PROMEDIO
# ==============================

rating1 = tienda1["Calificacion"].mean()
rating2 = tienda2["Calificacion"].mean()
rating3 = tienda3["Calificacion"].mean()
rating4 = tienda4["Calificacion"].mean()

print("\nCalificación promedio por tienda")
print("Tienda 1:", rating1)
print("Tienda 2:", rating2)
print("Tienda 3:", rating3)
print("Tienda 4:", rating4)

# ==============================
# COSTO PROMEDIO DE ENVÍO
# ==============================

envio1 = tienda1["Costo de envio"].mean()
envio2 = tienda2["Costo de envio"].mean()
envio3 = tienda3["Costo de envio"].mean()
envio4 = tienda4["Costo de envio"].mean()

print("\nCosto promedio de envío")
print("Tienda 1:", envio1)
print("Tienda 2:", envio2)
print("Tienda 3:", envio3)
print("Tienda 4:", envio4)

# ==============================
# GRÁFICO DE FACTURACIÓN
# ==============================

tiendas = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
facturacion = [facturacion1,facturacion2,facturacion3,facturacion4]

plt.figure(figsize=(8,5))
plt.bar(tiendas,facturacion)
plt.title("Facturación por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ventas")
plt.show()

# ==============================
# CONCLUSIÓN
# ==============================

print("\nConclusión del análisis:")
print("Después de analizar los datos de las tiendas se pueden observar diferencias en ventas, calificaciones y costos de envío.")
print("Algunas tiendas presentan mejores resultados en ventas y satisfacción de clientes.")
print("Este análisis permite comparar el rendimiento de cada tienda y apoyar la toma de decisiones basadas en datos.")
```
